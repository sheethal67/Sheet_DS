```python
X = Label_data.iloc[:,3:6]
y = Label_data.iloc[:,-1]
```

```python
train_dist = int(0.7 * len(X))
X_train, X_test = X[:train_dist], X[train_dist:]
y_train, y_test = y[:train_dist], y[train_dist:]
```

# Naive Bayes Classification Model

```python
Naive_model = GaussianNB()
```

```python
Naive_model.fit(X_train, y_train)
```

```python
Naive_predicted = Naive_model.predict(X_test)
```

```python
Naive_predicted
```

```python
Naive_count_misclassified = (y_test != Naive_predicted).sum()
print('Misclassified samples: {}'.format(Naive_count_misclassified))
```

```python
Naive_accuracy = metrics.accuracy_score(y_test, Naive_predicted)
print('Accuracy: {:.2f}'.format(Naive_accuracy))
```

```python
print(confusion_matrix(y_test, Naive_predicted))
```

```python
print(metrics.classification_report(y_test, Naive_predicted))
```
