```python
def bar_chart(feature):
    Survived = ttrain[ttrain['Survived']==1][feature].value_counts()
    Unsurvived = ttrain[ttrain['Survived']==0][feature].value_counts()
    Sex_data = pd.DataFrame([Survived, Unsurvived])
    Sex_data.index = ['Survived','Unsurvived']
    Sex_data.plot(kind='bar',stacked=True, figsize=(10,5))
```
